```python
import matplotlib.pyplot as plt
import networkx as nx

# Membuat grafik jaringan semantik yang lebih kaya
G_extended = nx.DiGraph()

# Menambahkan node (objek yang lebih kaya)
extended_nodes = [
    "Hewan",
    "Unggas",
    "Ayam",
    "Ayam Kampung",
    "Ayam Broiler",
    "Ayam Petelur",
    "Ayam Jantan",
    "Ayam Betina",
    "Burung Hantu",
    "Kalkun",
    "Merpati",
    "Cipung",
    "Penguin",
    "Sayap",
    "Kaki",
    "Ekor",
    "Bulu",
    "Biji-bijian",
    "Serangga",
    "Rodentia",
    "Ikan",
    "Kelinci",
    "Rumput",
    "Sayuran",
    "Buah-buahan",
    "Krill",
    "Gurita",
    "Kepiting",
    "Plankton",
    "Peternakan",
    "Kandang",
    "Ladang",
    "Pekan",
    "Kota",
    "Hutan",
    "Kawasan Terbuka",
    "Pegunungan",
    "Area Perkotaan",
    "Taman",
    "Antartika",
    "Pulau Terpencil",
    "Pantai Bebatuan",
    "Es Laut",
    "Daerah Dingin"
]

G_extended.add_nodes_from(extended_nodes)

# Menambahkan edges (hubungan antar objek yang lebih kaya)
extended_edges = [
    ("Hewan", "Unggas"),
    ("Unggas", "Ayam"),
    ("Unggas", "Burung Hantu"),
    ("Unggas", "Kalkun"),
    ("Unggas", "Merpati"),
    ("Unggas", "Cipung"),
    ("Unggas", "Penguin"),
    ("Ayam", "Kaki"),
    ("Ayam", "Paruh"),
    ("Ayam", "Ekor"),
    ("Ayam", "Bulu"),
    ("Ayam", "Biji-bijian"),
    ("Ayam", "Serangga"),
    ("Ayam", "Ayam Kampung"),
    ("Ayam", "Ayam Broiler"),
    ("Ayam", "Ayam Petelur"),
    ("Ayam", "Ayam Jantan"),
    ("Ayam", "Ayam Betina"),
    ("Burung Hantu", "Rodentia"),
    ("Burung Hantu", "Serangga"),
    ("Kalkun", "Biji-bijian"),
    ("Kalkun", "Sayuran"),
    ("Penguin", "Ikan"),
    ("Penguin", "Krill"),
    ("Habitat", "Peternakan"),
    ("Habitat", "Hutan"),
    ("Habitat", "Kawasan Terbuka"),
    ("Habitat", "Pegunungan"),
    ("Habitat", "Area Perkotaan"),
    ("Habitat", "Taman"),
    ("Habitat", "Antartika"),
    ("Habitat", "Pulau Terpencil"),
    ("Habitat", "Pantai Bebatuan"),
    ("Habitat", "Es Laut"),
    ("Habitat", "Daerah Dingin")
]

G_extended.add_edges_from(extended_edges)

# Posisi node dengan layout yang lebih terstruktur
pos_extended = nx.spring_layout(G_extended, seed=42, k=0.3)

# Gambar Jaringan
plt.figure(figsize=(16, 12))
nx.draw(G_extended, pos_extended, with_labels=True, node_size=3500, node_color="lightblue",
        font_size=12, font_color="black", font_weight="bold", edge_color="gray")

# Tandai node tertentu dengan warna merah
highlight_nodes_extended = ["Ayam", "Unggas", "Biji-bijian"]
nx.draw_networkx_nodes(G_extended, pos_extended, nodelist=highlight_nodes_extended, node_color="red", node_size=4000)

# Tambahkan keterangan untuk setiap relasi
edge_labels = {
    ("Unggas", "Ayam"): "ISA",
    ("Unggas", "Burung Hantu"): "ISA",
    ("Unggas", "Kalkun"): "ISA",
    ("Unggas", "Merpati"): "ISA",
    ("Ayam", "Biji-bijian"): "Makan",
    ("Ayam", "Kaki"): "HAS-PART",
    ("Burung Hantu", "Rodentia"): "Makan",
    ("Penguin", "Ikan"): "Makan",
    ("Habitat", "Peternakan"): "HABITAT",
    ("Habitat", "Hutan"): "HABITAT"
}

# Menambahkan keterangan di dalam grafik
nx.draw_networkx_edge_labels(G_extended, pos_extended, edge_labels=edge_labels, font_color='green', font_size=10)

plt.title("Jaringan Semantik yang Diperluas", fontsize=20)
plt.axis('off')
plt.show()
```

# Penjelasan Jaringan Semantik yang Diperluas

Jaringan semantik ini menggambarkan hubungan antara berbagai objek dalam kategori hewan, khususnya unggas, serta makanan dan habitat mereka. Berikut adalah penjelasan untuk setiap bagian dari jaringan:

## 1. Hewan
- **Hewan**: Merupakan kategori umum yang mencakup semua jenis hewan.
  - **Unggas**: Merupakan subkategori dari hewan yang mencakup spesies dengan sayap dan umumnya dapat terbang.

## 2. Unggas
- **Unggas**: Memiliki beberapa relasi dengan berbagai jenis unggas:
  - **Ayam**: Termasuk berbagai jenis ayam yang dibedakan berdasarkan fungsi dan sifatnya.
    - **Ayam Kampung**: Jenis ayam lokal yang biasanya dibesarkan di pekarangan.
    - **Ayam Broiler**: Jenis ayam yang dibesarkan untuk diambil dagingnya.
    - **Ayam Petelur**: Jenis ayam yang dibesarkan untuk diambil telurnya.
    - **Ayam Jantan**: Jenis ayam jantan yang biasanya lebih besar dan memiliki warna cerah.
    - **Ayam Betina**: Jenis ayam betina yang berfungsi untuk bertelur.
  
  - **Burung Hantu**: Jenis unggas yang aktif di malam hari dan terkenal sebagai predator.
    - **Rodentia**: Makanan utama burung hantu, termasuk tikus.
    - **Serangga**: Juga merupakan bagian dari diet burung hantu.

  - **Kalkun**: Unggas besar yang sering dijadikan makanan pada perayaan tertentu.
    - **Biji-bijian**: Salah satu makanan kalkun.
    - **Sayuran**: Dapat menjadi bagian dari diet kalkun.

  - **Merpati**: Unggas yang sering ditemukan di lingkungan urban.
  - **Cipung**: Jenis burung kecil yang juga termasuk dalam kategori unggas.
  - **Penguin**: Unggas yang tidak bisa terbang dan biasanya ditemukan di daerah dingin.
    - **Ikan**: Makanan utama penguin.
    - **Krill**: Sumber makanan lain bagi penguin.

## 3. Sayap
- **Sayap**: Merupakan bagian penting dari unggas yang membantu mereka terbang. Juga memiliki relasi sebagai bagian dari tubuh unggas.
  - **Kaki**: Digunakan untuk bergerak dan mencari makanan.
  - **Ekor**: Membantu dalam keseimbangan saat terbang.
  - **Bulu**: Menyediakan isolasi dan membantu dalam penerbangan.

## 4. Makanan
- **Biji-bijian**: Merupakan makanan umum bagi banyak unggas, termasuk ayam dan kalkun.
- **Serangga**: Bagian penting dari diet burung hantu dan ayam.
- **Sayuran**: Juga sering dimakan oleh ayam dan kalkun.
- **Buah-buahan**: Merupakan bagian dari diet unggas lainnya.
- **Ikan**: Makanan utama bagi penguin.

## 5. Habitat
- **Habitat**: Menunjukkan tempat tinggal berbagai jenis unggas.
  - **Peternakan**: Tempat di mana ayam biasanya dibesarkan.
  - **Hutan**: Habitat alami untuk burung hantu dan berbagai unggas lainnya.
  - **Kawasan Terbuka**: Lingkungan yang juga dapat didiami oleh berbagai spesies unggas.
  - **Pegunungan**: Habitat untuk unggas tertentu.
  - **Area Perkotaan**: Tempat tinggal untuk merpati dan unggas lain yang beradaptasi dengan lingkungan urban.
  - **Antartika**: Habitat alami bagi penguin.
  - **Pulau Terpencil**: Tempat tinggal bagi berbagai spesies yang terisolasi.

## Kesimpulan
Jaringan semantik ini mencerminkan kompleksitas dan keanekaragaman dalam ekosistem unggas, menunjukkan bagaimana berbagai spesies saling terkait melalui makanan, habitat, dan relasi biologis lainnya.

```python
# Definisikan struktur Frame menggunakan Dictionary
class Frame:
    def __init__(self, name):
        self.name = name
        self.slots = {}
        self.explanation = ""

    # Menambahkan slot ke frame
    def add_slot(self, slot_name, filler):
        self.slots[slot_name] = filler

    # Menambahkan penjelasan ke frame
    def add_explanation(self, explanation):
        self.explanation = explanation

    # Menampilkan informasi dari frame
    def show(self):
        print(f"Frame: {self.name}")
        for slot, filler in self.slots.items():
            print(f"  Slot: {slot}  {filler}")
        print()

    # Menampilkan penjelasan dari frame
    def explain(self):
        print(f"Explanation for {self.name}: {self.explanation}\n")

# Definisikan Semantic Network sebagai Frames

# Frame untuk Animal
animal_frame = Frame("Animal")
animal_frame.add_slot("ISA", "LivingBeing")
animal_frame.add_explanation("Animal adalah makhluk hidup. Frame ini menggambarkan kategori umum dari makhluk hidup yang termasuk hewan.")

# Frame untuk Mammal
mammal_frame = Frame("Mammal")
mammal_frame.add_slot("ISA", "Animal")
mammal_frame.add_slot("HAS-PART", "tail")
mammal_frame.add_explanation("Mammal adalah subkategori dari Animal, dan mamalia memiliki bagian tubuh berupa ekor (tail).")

# Frame untuk Cat
cat_frame = Frame("Cat")
cat_frame.add_slot("ISA", "Mammal")
cat_frame.add_slot("HAS-PART", "tail")
cat_frame.add_slot("EATS", "Cat food")
cat_frame.add_slot("INST", ["Garfield", "Sylvester"])
cat_frame.add_explanation("Cat adalah mamalia dan memiliki bagian ekor serta memakan makanan kucing. Contoh kucing termasuk Garfield dan Sylvester.")

# Frame untuk Garfield
garfield_frame = Frame("Garfield")
garfield_frame.add_slot("ISA", "Cat")
garfield_frame.add_slot("EATS", "Lasagna")
garfield_frame.add_explanation("Garfield adalah contoh dari kucing yang terkenal dan memiliki kebiasaan makan lasagna.")

# Frame untuk Sylvester
sylvester_frame = Frame("Sylvester")
sylvester_frame.add_slot("ISA", "Cat")
sylvester_frame.add_explanation("Sylvester adalah kucing lain yang termasuk dalam kategori Cat.")

# Frame untuk Bird
bird_frame = Frame("Bird")
bird_frame.add_slot("ISA", "Animal")
bird_frame.add_slot("HAS-PART", ["wing", "feather"])
bird_frame.add_slot("EATS", "Seeds")
bird_frame.add_slot("CAN", "Fly")
bird_frame.add_explanation("Bird adalah hewan yang memiliki sayap dan bulu, makan biji-bijian, serta mampu terbang.")

# Frame untuk Canary
canary_frame = Frame("Canary")
canary_frame.add_slot("ISA", "Bird")
canary_frame.add_slot("EATS", "Seeds")
canary_frame.add_slot("INST", "Tweety")
canary_frame.add_explanation("Canary adalah sejenis burung yang memakan biji-bijian. Tweety adalah contoh dari burung kenari.")

# Frame untuk Penguin
penguin_frame = Frame("Penguin")
penguin_frame.add_slot("ISA", "Bird")
penguin_frame.add_slot("CANNOT", "Fly")
penguin_frame.add_slot("INST", "Opus")
penguin_frame.add_explanation("Penguin adalah burung yang tidak bisa terbang. Contoh penguin adalah Opus.")

# Frame untuk Tweety
tweety_frame = Frame("Tweety")
tweety_frame.add_slot("ISA", "Canary")
tweety_frame.add_explanation("Tweety adalah contoh spesifik dari burung kenari.")

# Frame untuk Opus
opus_frame = Frame("Opus")
opus_frame.add_slot("ISA", "Penguin")
opus_frame.add_explanation("Opus adalah penguin, yang merupakan burung yang tidak bisa terbang.")

# Menampilkan semua frame beserta penjelasannya
frames = [animal_frame, mammal_frame, cat_frame, garfield_frame, sylvester_frame, bird_frame, canary_frame, penguin_frame, tweety_frame, opus_frame]

# Tampilkan frame
for frame in frames:
    frame.show()
    frame.explain()
```
